```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from causallib.estimation import IPW, AIPW, StratifiedStandardization

# Defining treatment and outcome
treatment = defenders['IsTackler']  # Treatment: Is Tackler (binary)
outcome = defenders['expectedPointsAdded']  # Outcome: EPA (continuous)

# Features to be used by both the treatment and outcome models
features = ['x', 'y', 's', 'a', 'o', 'dir', 'ballcarrier_o', 'ballcarrier_dir', 'ballcarrier_dis',
            'ballcarrier_x', 'ballcarrier_y', 'ballcarrier_s', 'ballcarrier_a', 'dis_to_ballcarrier',
            'orientation_diff_to_ballcarrier', 'direction_diff_to_ballcarrier', 'opp1Dist', 'opp1X', 'opp1Y',
            'opp1Dir', 'opp1O', 'opp1S', 'opp1A', 'opp2Dist', 'opp2X', 'opp2Y', 'opp2Dir', 'opp2O', 'opp2S', 'opp2A',
            'opp3Dist', 'opp3X', 'opp3Y', 'opp3Dir', 'opp3O', 'opp3S', 'opp4Dist', 'opp4X', 'opp4Y', 'opp4Dir',
            'opp4O', 'opp4S', 'opp4A', 'opp5Dist', 'opp5X', 'opp5Y', 'opp5Dir', 'opp5O', 'opp5S', 'opp5A', 'opp6Dist',
            'opp6X', 'opp6Y', 'opp6Dir', 'opp6O', 'opp6S', 'opp6A', 'opp7Dist', 'opp7X', 'opp7Y', 'opp7Dir', 'opp7O',
            'opp7S', 'opp7A', 'opp8Dist', 'opp8X', 'opp8Y', 'opp8Dir', 'opp8O', 'opp8S', 'opp8A', 'opp9Dist', 'opp9X',
            'opp9Y', 'opp9Dir', 'opp9O', 'opp9S', 'mate1Dist', 'mate1X', 'mate1Y', 'mate1Dir', 'mate1O', 'mate1S', 
            'mate1A', 'mate2Dist', 'mate2X', 'mate2Y', 'mate2Dir', 'mate2O', 'mate2S', 'mate2A', 'mate3Dist', 
            'mate3X', 'mate3Y', 'mate3Dir', 'mate3O', 'mate3S', 'mate3A', 'mate4Dist', 'mate4X', 'mate4Y', 'mate4Dir', 
            'mate4O', 'mate4S', 'mate4A', 'mate5Dist', 'mate5X', 'mate5Y', 'mate5Dir', 'mate5O', 'mate5S', 'mate5A', 
            'mate6Dist', 'mate6X', 'mate6Y', 'mate6Dir', 'mate6O', 'mate6S', 'mate6A', 'mate7Dist', 'mate7X', 
            'mate7Y', 'mate7Dir', 'mate7O', 'mate7S', 'mate7A', 'mate8Dist', 'mate8X', 'mate8Y', 'mate8Dir', 'mate8O', 
            'mate8S', 'mate8A', 'mate9Dist', 'mate9X', 'mate9Y', 'mate9Dir', 'mate9O', 'mate9S']

# Step 1: Propensity Score Modeling using Gradient Boosting Classifier for binary treatment
# IPW estimator to handle treatment assignment
ipw = IPW(GradientBoostingClassifier(n_estimators=10, random_state=42, verbose=1))
ipw.fit(defenders[features], treatment)

# Step 2: Outcome Modeling using Gradient Boosting Regressor for continuous outcome (EPA)
# Stratified Standardization for modeling the continuous outcome (EPA)
outcome_model = StratifiedStandardization(GradientBoostingRegressor(n_estimators=10, random_state=42, verbose=1))
outcome_model.fit(defenders[features], treatment, outcome)

# Step 3: AIPW (Augmented Inverse Probability Weighting)
aipw_model = AIPW(outcome_model, ipw)

# Step 4: Fit the AIPW model using both the treatment and outcome models
aipw_model.fit(defenders[features], treatment, outcome)

# Output the results
print("AIPW model fitted successfully with Gradient Boosting Classifier and Regressor.")
```

```python
from causallib.evaluation import evaluate

results = evaluate(aipw_model.weight_model, defenders[features], treatment, outcome, cv='auto')
results.plot_all()
results.evaluated_metrics.prediction_scores
```

```python
fig, [a0, a1] = plt.subplots(1, 2, figsize=(12, 6))
results.plot_roc_curve(ax=a0)
results.plot_roc_curve(phase="valid", ax=a1)
for ax, suffix in [(a0, "Train"), (a1 ,"Validation")]:
    ax.set_title(ax.get_title() + ": " + suffix)
plt.tight_layout();

fig, [a0, a1] = plt.subplots(1, 2, figsize=(12, 6))
results.plot_calibration_curve(ax=a0)
results.plot_calibration_curve(phase="valid", ax=a1)
for ax, suffix in [(a0, "Train"), (a1 ,"Validation")]:
    ax.set_title(ax.get_title() + ": " + suffix)
plt.tight_layout();
```

```python
# Filter only rows where the player is the tackler
tackler_defenders = defenders[defenders['IsTackler'] == 1]

# Drop duplicate rows based on 'displayName', 'nflId', 'playId', and 'gameId'
tackler_defenders = tackler_defenders.drop_duplicates(subset=['displayName', 'nflId', 'playId', 'gameId'])

# Display the DataFrame
tackler_defenders
```

```python
# Group by 'displayName', 'position', and 'nflId', then calculate the mean and sum of 'CATE' and 'EPA_Tackle'
cate_tackle_aggregated = tackler_defenders.groupby(['displayName', 'position', 'nflId']).agg({
    'CATE': ['mean', 'sum'],  # Aggregating CATE by both mean and sum
    'EPA_Tackle': ['mean', 'sum'],  # Aggregating EPA_Tackle by both mean and sum
    'expectedPointsAdded': ['mean', 'sum']  # Aggregating expectedPointsAdded by both mean and sum
}).reset_index()

# Flattening the multi-level column names
cate_tackle_aggregated.columns = ['displayName', 'position', 'nflId', 'Avg_CATE_TackleMade', 'Sum_CATE_TackleMade', 'Avg_EPA_TackleMade', 'Sum_EPA_TackleMade', 'Mean_expectedPointsAdded_TackleMade', 'Sum_expectedPointsAdded_TackleMade']

# Display the aggregated DataFrame
cate_tackle_aggregated
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Filter out rows where 'Mean_expectedPointsAdded_TackleMade' is below -2
filtered_df = cate_tackle_aggregated[cate_tackle_aggregated['Mean_expectedPointsAdded_TackleMade'] >= -2]

# Extract the necessary columns from the filtered DataFrame
x = filtered_df['Avg_EPA_TackleMade']
y = filtered_df['Mean_expectedPointsAdded_TackleMade']

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7, edgecolors='b', label='Data Points')

# Fit a line of best fit
slope, intercept, r_value, p_value, std_err = linregress(x, y)
plt.plot(x, slope * x + intercept, color='red', label=f'Best Fit Line (r={r_value:.2f})')

# Calculate R-squared value
r_squared = r_value ** 2

# Add labels and title
plt.xlabel('Avg EPA Tackle')
plt.ylabel('Mean Expected Points Added')
plt.title(f'Scatter Plot of Avg EPA Tackle vs Mean Expected Points Added\n R²={r_squared:.2f}')

# Show correlation in the plot
plt.legend()

# Show plot
plt.show()

# Print correlation and R²
print(f"Correlation coefficient (r): {r_value:.2f}")
print(f"R² (coefficient of determination): {r_squared:.2f}")
```

```python
tackles = pd.read_csv('tackles.csv')
tackles
```

```python
player_agg_stats = tackles.groupby('nflId').agg(
    Tackle_Count=('tackle', 'sum'),  # Sum of tackles
    Assist_Count=('assist', 'sum'),  # Sum of assists
    Forced_Fumble_Count=('forcedFumble', 'sum'),  # Sum of forced fumbles
    Missed_Tackle_Count=('pff_missedTackle', 'sum'),  # Sum of missed tackles
    Unique_Tackles=('playId', 'nunique')  # Count of unique tackles (unique playIds)
).reset_index()

player_agg_stats
```

```python
# Perform the join with suffixes to handle overlapping column names
cate = cate_aggregated.join(player_agg_stats.set_index('nflId'), on='nflId', how='inner', lsuffix='_cate', rsuffix='_player')

# Display the result
cate
```

```python
cate['Tackles'] = cate['Tackle_Count'] + cate['Assist_Count']
cate
```
